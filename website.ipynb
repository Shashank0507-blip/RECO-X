```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
from sklearn.manifold import TSNE

## Load the dataset 
df = pd.read_csv("website_dataset.csv")

## Create a body double of the original
df1 = df.copy()

# Drop unnecessary columns
df1 = df1[['Website','Category']]
df1.head()

## define variable "X"
X = df1[['Category']]

## Encode the "category" column before performing clustering and convert it to an array.
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X).toarray()

## Apply K-means algorithm
kmeans = KMeans(n_clusters= 10, n_init='auto', random_state=42)
df1["Cluster"] = kmeans.fit_predict(X_encoded)
#--------------------------------------------------------------
import requests

def get_trending_movies():
    api_key = "a0b63f832267041be76f13ce6d9eea36"
    url = f"https://api.themoviedb.org/3/trending/movie/day?api_key={api_key}"
    response = requests.get(url)
    data = response.json()
    movies = [movie["title"] for movie in data["results"][:10]]
    return movies

def get_trending_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"  
    url = f"https://newsapi.org/v2/top-headlines?country=us&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    news = [article["title"] for article in data["articles"][:10]]
    return news

def get_trending_games():
    api_key = "ab4563fc0d1b445cbb5dd1a77dd0a284"  
    url = f"https://api.rawg.io/api/games?key={api_key}&ordering=-added&page_size=5"
    response = requests.get(url)
    data = response.json()
    games = [game["name"] for game in data["results"][:10]]
    return games

def get_tech_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590" 
    url = f"https://newsapi.org/v2/top-headlines?category=technology&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    tech_news = [article["title"] for article in data["articles"][:10]]
    return tech_news

def get_trending_health_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"
    url = f"https://newsapi.org/v2/top-headlines?country=us&category=health&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    health_news = [article["title"] for article in data["articles"][:10]]
    return health_news

def get_trending_social_media_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"
    url = f"https://newsapi.org/v2/everything?q=social%20media&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    social_media_news = [article["title"] for article in data["articles"][:10]]
    return social_media_news

def get_trending_finance_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"
    url = f"https://newsapi.org/v2/top-headlines?country=us&category=business&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    finance_news = [article["title"] for article in data["articles"][:10]]
    return finance_news

def get_trending_education_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"
    url = f"https://newsapi.org/v2/everything?q=education&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    education_news = [article["title"] for article in data["articles"][:10]]
    return education_news

def get_trending_music_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"
    url = f"https://newsapi.org/v2/everything?q=music&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    music_news = [article["title"] for article in data["articles"][:10]]
    return music_news

def get_trending_shopping_news():
    api_key = "ed7cef9d0fef4079a56f7cec4307a590"
    url = f"https://newsapi.org/v2/everything?q=shopping&apiKey={api_key}"
    response = requests.get(url)
    data = response.json()
    shopping_news = [article["title"] for article in data["articles"][:10]]
    return shopping_news

#----------------------------------------------------------------------
def suggest_websites():
    user_category = input("What are you in the mood for today ðŸ˜‰ðŸ˜‰: ").strip()

    # Find the category in the dataset
    if user_category in df1['Category'].values:
        cluster = df1[df1['Category'] == user_category]['Cluster'].iloc[0]
        
        # Retrieve other websites from the same cluster
        similar_websites = df1[df1['Cluster'] == cluster]['Website'].tolist()

        # Display recommendations
        print(f"\n Top Platforms for '{user_category}':")
        for web in similar_websites[:10]:  # Show up to 10 recommendations
            print(f"- {web} ")
        # Fetch trending movies if the category = "Entertainment [not the kind u are looking for:) ]"
        if user_category.lower() == "entertainment":
            print("\nTrending Movies Right Now:")
            trending_movies = get_trending_movies()
            for movie in trending_movies:
                print(f"- {movie}")
        # Fetch trending news if the category = "News"
        if user_category.lower() == "news":
            print("\nHot News Right Now:")
            trending_news = get_trending_news()
            for news in trending_news:
                print(f"- {news}")
        # Fetch trending games if the category = "games"
        if user_category.lower() == "games":
            print("\nHot games Right Now:")
            trending_games = get_trending_games()
            for game in trending_games:
                print(f"- {game}")
        # Fetch trending news about tech if the category = "technology"
        if user_category.lower() == "technology":
            print("\nHot tech news  Right Now ðŸ¤–:")
            trending_tech_news = get_tech_news()
            for tech_news in trending_tech_news:
                print(f"- {tech_news}")
        # Fetch trending news about health if the category = "health"
        if user_category.lower() == "health":
            print("\nHealthy News: (is that even a thingðŸ¤”) :")
            trending_health_news = get_trending_health_news()
            for health_news in trending_health_news:
                print(f"- {health_news}")
        # Fetch trending news about social media if the category = "social media"
        if user_category.lower() == "social media":
            print("\nSocial Media News:(The things that affect maximum 2 peopleðŸ˜‚) :")
            trending_social_media_news = get_trending_social_media_news()
            for social_news in trending_social_media_news:
                print(f"- {social_news}")
        # Fetch trending news about social media if the category = "social media"
        if user_category.lower() == "finance":
            print("\nFinancial News ðŸ“ˆ :")
            trending_money_news = get_trending_finance_news()
            for finance_news in trending_money_news:
                print(f"- {finance_news}")
        # Fetch trending news about Education if the category = "Education"
        if user_category.lower() == "education":
            print("\nEducation news to pick your future:")
            trending_edu_news = get_trending_education_news()
            for edu_news in trending_edu_news:
                print(f"- {edu_news}")
        # Fetch trending news about shopping if the category = "shopping"
        if user_category.lower() == "shopping":
            print("\nShopping News (be updatedðŸ¤¯):")
            trending_shop_news = get_trending_shopping_news()
            for shop_news in trending_shop_news:
                print(f"- {shop_news}")
        # Fetch trending news about music if the category = "music"
        if user_category.lower() == "music":
            print("\nMusic Stuff ðŸŽ¸ðŸŽ¸:")
            trending_music_news = get_trending_music_news()
            for music_news in trending_music_news:
                print(f"- {music_news}")
    else:
        print(" Not found in the dataset. Try again!")

# Call the function to get user input and suggest similar movies
suggest_websites()
```
